```python
# Part 1

def match_next(dx, dy, new_pipe):

    match dx:
        case -1:
            match new_pipe:
                case 'L':
                    return (0,-1)
                case '-':
                    return (-1,0)
                case 'F':
                    return (0,1)
        case 1:
            match new_pipe:
                case '-':
                    return (1,0)
                case 'J':
                    return (0,-1)
                case '7':
                    return (0,1)
        case 0:
            match dy:
                case 1:
                    match new_pipe:
                        case '|':
                            return (0,-1)
                        case 'L':
                            return (1,0)
                        case 'J':
                            return (-1,0)
                case -1:
                    match new_pipe:
                        case '|':
                            return (0,1)
                        case 'F':
                            return (1,0)
                        case '7':
                            return (-1,0)
    return (-1,-1)

from collections import deque

file = open('day_10_part_1.txt', 'r')

lines = file.readlines()


DIRECTIONS = [(1,0), (-1,0), (0,1), (0,-1)]

q = deque()
moves = 0

# In the queue we store the position's coordinates, and the previously-visited coordinates to prevent backtracking
for row, line in enumerate(lines):
    lines[row] = lines[row].strip()
    if 'S' in line:
        q.append((line.index('S'), row, '*')) # column then row, i.e. x then y
        
M, N = len(lines), len(lines[0])
visited = [[False] * N for _ in range(M)]

while q:
    for _ in range(len(q)):

        x, y, prev = q.popleft()

        for dx, dy in DIRECTIONS:
            if (x+dx in range(N) and
            (y+dy) in range(M) and
            not visited[y+dy][x+dx] and
            (x+dx, y+dy) != prev):
                new_next = match_next(dx,dy,lines[y+dy][x+dx])
                if new_next != (-1,-1):
                    q.append((x+dx, y+dy, (x, y)))

        visited[y][x] = True
    moves += 1

print(moves - 1)

#6864
```
